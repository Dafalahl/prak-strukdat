```python
from collections import deque

class TowerOfHanoi:
    def __init__(self, max_moves=8):
        # Inisialisasi tiga tiang dengan stack (deque)
        self.towers = {
            "A": deque(["游댮", "游", "游리", "游댯"]),  # Semua cakram awal di tiang A
            "B": deque(),
            "C": deque()
        }
        self.moves = 0
        self.max_moves = 8

    def move(self, source, target):
        """Memindahkan cakram dari satu tiang ke tiang lain."""
        if self.moves >= self.max_moves:
            print("Batas maksimal gerakan telah tercapai!")
            return

        if self.towers[source]:
            disk = self.towers[source].pop()
            self.towers[target].append(disk)
            self.moves += 1  # Tambah jumlah gerakan
            self.print_towers()
        else:
            print(f"Tidak ada cakram di {source} untuk dipindahkan!")

    def print_towers(self):
        """Menampilkan kondisi terkini tiang."""
        print(f"A: {' '.join(self.towers['A'])}")
        print(f"B: {' '.join(self.towers['B'])}")
        print(f"C: {' '.join(self.towers['C'])}")

# Buat instance permainan dengan batas 8 gerakan
game = TowerOfHanoi()
game.print_towers()

# Memainkan game secara interaktif
while True:
    if game.moves >= game.max_moves:
        print("Permainan selesai! Anda telah mencapai batas gerakan.")
        break

    source = input("Masukkan tiang asal (A/B/C) atau 'exit' untuk keluar: ").upper()
    if source == "EXIT":
        print("Permainan selesai!")
        break

    target = input("Masukkan tiang tujuan (A/B/C): ").upper()

    if source in game.towers and target in game.towers:
        game.move(source, target)
    else:
        print("Input tidak valid! Coba lagi.")
```
